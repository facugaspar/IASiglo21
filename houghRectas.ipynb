<h3>1. Se importan las librerias a Utilizar</h3>

```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

<h3>2. Se lee la imagen a evaluar</h3>

```python
img = cv.imread('perforacion.png')

plt.imshow(img)
plt.show()
```

<h3>3. Se binariza la imagen para para detectar los bordes</h3>

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
edges = cv.Canny(gray, 50, 150, apertureSize = 3)

plt.imshow(edges)
plt.show()
```

<h3>5. Se aplica la Transformada de Hough</h3>

```python
lines = cv.HoughLinesP(edges, 1, np.pi/180, 50, minLineLength=1, maxLineGap=10)

for line in lines:
    x1, y1, x2, y2 = line[0]
    cv.line(img, (x1,y1), (x2,y2), (0,255,0), 1, cv.LINE_AA)
```

<h3>6. Se imprimen en color Verde las rectas encontradas</h3>

```python

plt.imshow(img)
plt.show()
```
